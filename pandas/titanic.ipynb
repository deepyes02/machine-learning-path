```python
import pandas as pd
# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
# Display the first few rows of the dataset
df.head()
```

```python
import torch
from sklearn.model_selection import train_test_split
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
df
```

```python
####Basically, we fill the not available data with median for age and mdoe for embarked
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()) # Q - Queenstown, C - Cherbourg, S - Southhampton
## sex - Male is 0 and female is 1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
# change Embarked and PClass to boolean columns
df = pd.get_dummies(df, columns=['Embarked', 'Pclass'], drop_first=True)
df
```
