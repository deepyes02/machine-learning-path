```python
import matplotlib.pyplot as plt
import numpy as np

# Data
score_intervals = ['0-2', '3-5', '6-8', '9-10']
frequencies = [3, 8, 7, 2]
midpoints = [1, 4, 7, 9.5]
mean = 5.15
deviations = [abs(midpoint - mean) for midpoint in midpoints]
print(deviations)
normalized_deviations = [deviation / max(deviations) for deviation in deviations]

# Plotting
fig, ax1 = plt.subplots()

# Plotting the frequencies
ax1.bar(score_intervals, frequencies, color='blue', alpha=0.6, label='Frequency')
ax1.set_xlabel('Score Interval')
ax1.set_ylabel('Frequency', color='red')
ax1.tick_params(axis='y', labelcolor='green')

# Creating a second y-axis for deviations
ax2 = ax1.twinx()
ax2.plot(score_intervals, normalized_deviations, color='black', marker='o', label='Normalized Deviation')
ax2.set_ylabel('Normalized Deviation', color='black')
ax2.tick_params(axis='y', labelcolor='brown')

# Adding a legend
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(1, 0.9))

plt.title('Frequency and Normalized Deviation of Score Intervals')
plt.show()
```
